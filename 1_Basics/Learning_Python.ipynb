```python
from datasets import load_dataset
```

```python
datasets = load_dataset('lukebarousse/data_jobs')
```

```python
df = datasets['train'].to_pandas()
```

```python
df[['job_title_short', 'job_title']].iloc[16:23]
```

```python
df['job_title_short'].unique()
```

```python
df[(df.job_title_short == 'Data Engineer') & (df.job_work_from_home == True) & (df.salary_year_avg.notna())
 & (df.salary_rate.notna())]
```

```python
df.info()
```

```python
df.job_posted_date = pd.to_datetime(df.job_posted_date)
```

```python
df.info()
```

```python
df['job_posted_month'] = df.job_posted_date.dt.month
```

```python
df.drop('salary_hour_avg', axis = 1, inplace= True )
```

```python
df.info()
```

```python
df.dropna(subset = 'salary_year_avg', inplace = True)
```

```python
df.groupby("job_title_short")['salary_year_avg'].agg(['min', 'max', 'median'])
```

```python
india_jobs = df[df['job_country']== "India"]
```

```python
 india_jobs
```

```python
india_jobs.groupby(["job_title_short"])["salary_year_avg"].agg(['min', 'max', 'mean', 'count', 'median']).sort_values(by= "median", ascending = False)
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
monthly_counts = df.job_posted_month.value_counts()
monthly_counts = monthly_counts.sort_index()
monthly_counts
```

```python
plt.plot(monthly_counts.index , monthly_counts)
```

```python
title_counts = df.job_title_short.value_counts()
title_counts = title_counts.sort_values(ascending = False)
```

```python
plt.bar(title_counts.index, title_counts)
plt.title('No. of postings by Job title')
plt.ylabel('Count of job postings')
plt.xticks(rotation = 45, ha= 'right')
plt.show()
```

```python
title_counts.plot(kind= 'bar')
plt.title('No. of postings by title')
plt.ylabel('count of job postings')
plt.xlabel("")
plt.xticks(rotation = 43 , ha = 'right')
plt.show()
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
```

```python
job_salary.plot(kind = 'barh')
plt.title('Median salary by Job title')
plt.xlabel('USD$')
plt.ylabel("")
plt.show()
```
